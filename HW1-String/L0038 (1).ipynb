```python
############################################################
# Solution.py
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2020
###########################################################
class Solution:
  def __init__(self):
    pass

  #THIS IS LEETCODE
  def  countAndSay(self,n:'int')->'string':
    ## NOTHING CAN BE CHANGED HERE
    return self._alg_integer(n)

  def  countAndSayString(self,n:'string')->'string':
    #Nothing can be changed here
    return self._alg_string(n) 

  ### YOU CAN HAVE ANY NUMBER OF FUNCTIONS AND CLASSES BELOW


  ## n is str here
  def _alg_string(self,n:'string')->'string':
    print("WRITE CODE  BELOW")
    newstring=""
    pa, pb=0,1 
    while(pb<len(n)):
        if(n[pa]==n[pb]):
            pb+=1
        else:
            newstring+=str(pb-pa)+n[pa]
            pa=pb
            pb+=1
            
    newstring+=str(pb-pa)+n[pa]
    
    return newstring

  ## n is int here
  def _alg_integer(self,n:'int')->'string':
    print("WRITE CODE  BELOW")
    string="1"

    if(n==1):
        return string

    for i in range(1,n):
        newstring=""

        pa, pb = 0, 1
        while(pb<len(string)):
            if(string[pa]==string[pb]):
                pb+=1
            else:
                newstring+=str(pb-pa)+string[pa]
                pa=pb
                pb+=1
            
        newstring+=str(pb-pa)+string[pa]
        string=newstring
        
    return string
```
