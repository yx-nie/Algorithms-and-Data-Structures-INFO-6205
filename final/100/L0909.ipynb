```python
############################################################
# Exam.py
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2023
###########################################################

############################################################
# All imports
###########################################################
from typing import List
from time import process_time 
import collections
import math

def xxxx(ladders, snakes):
    num_moves = [-1]
    work = [0]
    max = 100
    p = Exam(max,ladders, snakes,num_moves,work,False)
    return num_moves[0]

    
class Exam:
    def __init__(self, max:'int', ladders:'list', snakes:'list', num_moves:'List of size 1',work:'List of size 1',show:'boolean'):
        self._max = max #size of board
        self._ladders = ladders
        self._snakes = snakes
        self._num_moves = num_moves
        self._work = work
        self._show = show

        ## you can initialize all your private variables here STARTS
        self._boardsize=int(math.sqrt(self._max))
        self._queue=collections.deque()
        self._queue.append((1,[1]))
        self._visited=set()
        self._shortestpath={}

        
	    ## you can initialize all your private variables here ENDS
        self._solution()


    def _set_num_moves(self, m:'int'):
        self._num_moves[0] = m 

    def _increment_work(self):
        self._work[0] = self._work[0] + 1

    def _solution(self):
        #MUST WRITE 2 routines
        self._alg()
        if (self._show):
            # 2) Print shortest path
            #ONLY 50 marks will be given if you cannot print the shortest path
            print("The paths from 1 to",self._max, "are as follows")
            self._print_all_shortest_path()
            print("WORK DONE =",self._work[0])

    ############################################################
    # WRITE CODE BELOW
    ###########################################################
              
    def _alg(self):
        print("WRITE CODE")

        while(self._queue):
            current_pos, current_path=self._queue.popleft()
            self._visited.add(current_pos)
                
            if (current_pos==self._max):
                self._set_num_moves(len(current_path)-1)
                return
            ## reach the end pooint of the baord
            ##other wise, keep adding to the queue

            ##from 0 to self._boardsize, iterate check ladder or snake
            for j in range(1,self._boardsize+1):
                self._increment_work()
                nextpos=current_pos+j
                

                if(nextpos <= self._max):
                    ##check nextpos is in the ladder or snake or not
                    ladder_pos = self._checkLadder(nextpos, self._ladders)
                    snake_pos = self._checkSnake(nextpos, self._snakes)

                    ## if nextpos is in the ladder or in the snake
                    ## current path should be terminated, and pass the path to the nextpost. and the current path length is set 0
                    ## will consider it in the printout.

                    ## if in, cover the path to the nextpos.
                    if ladder_pos>0:
                        nextpos = ladder_pos
                    elif snake_pos>0:
                        nextpos = snake_pos
                    
                    ## if nexpost not visited, put into the queue for next iteration
                    ###makessuer each number is iterated
                    if (nextpos not in self._visited):
                        new_path = current_path.copy()
                        new_path.append(nextpos)
                        self._queue.append((nextpos, new_path))
                        

                    #if not stored or not the shorteste one, update it
                    if nextpos not in self._shortestpath or len(new_path) < len(self._shortestpath[nextpos]):
                        self._shortestpath[nextpos] = new_path




    def _print_all_shortest_path(self):
        print("WRITE CODE")
        ##consider the cover number like 3
        for i in range(1, self._max+1):
            if i in self._shortestpath:
                self._print_path(self._shortestpath[i])
            else:
                if (i > 1):
                    print(1, " -> ", i, " NO Path")
    
    def _checkLadder(self, pos:'int', ladder:'list'):
        for i in range(len(ladder)):
            self._increment_work()
            if (pos == ladder[i][0]):
                return ladder[i][1]
        return -1
    
    def _checkSnake(self, pos:'int', snake:'list'):
        for i in range(len(snake)):
            self._increment_work()
            if (pos == snake[i][0]):
                return snake[i][1]
        return -1
    
    def _print_path(self, path):
        print(" -> ".join(map(str, path)),"Length=", len(path)-1)

     


```
